# Fichier des médecins par département

```python
unames = ['Specialite','Dept','nb_medic','h_ss_dep', 'h_dep', 'h_frais_dep', 'h_total']
file = '/Users/matthieuroussel/OneDrive/Documents/SauvegardeMac/Documents/Telecom Paristech Enseignements/MDI721 - Kit DataScience/Lesson 5/datas/Honoraires_totaux_des_professionnels_de_sante_par_departement_en_2016-2.xls'
df_med1 = pd.read_excel(file, sheet_name="Spécialistes",header=0, na_values="nc", names = unames)
df_med1['Type'] = 'Spécialistes'
```

```python
df_med2 = pd.read_excel(file, sheet_name="Généralistes et MEP",header=0, na_values="nc", names = unames)
df_med2['Type'] = 'Généralistes et MEP'
```

```python
df_med3 = pd.read_excel(file, sheet_name="Dentistes et ODF",header=0, na_values="nc", names = unames)
df_med3['Type'] = 'Dentistes et ODF'
```

```python
df_med4 = pd.read_excel(file, sheet_name="Sages-femmes",header=0, na_values="nc", names = unames)
df_med4['Type'] = 'Sages-femmes'
```

```python
df_med5 = pd.read_excel(file, sheet_name="Auxiliaires médicaux",header=0, na_values="nc", names = unames)
df_med5['Type'] = 'Auxiliaires médicaux'
```

```python
df_med6 = pd.read_excel(file, sheet_name="Laboratoires",header=0, na_values="nc", names = unames)
df_med6['Type'] = 'Laboratoires'
```

```python
frames = [df_med1, df_med2, df_med3,df_med4,df_med5,df_med6]
```

```python
df_final = pd.concat(frames, ignore_index=True)
```

```python
patternDel = "TOTAL+"
filter = df_final['Specialite'].str.contains(patternDel)
```

```python
df_final = df_final[~filter]
```

```python
df_final.groupby(['Type','Specialite']).size()
```

```python
reg3 = r'(\d+\w)'
df_final['num_dept'] = df_final['Dept'].str.extract(reg3, expand=False)
```

```python
df_med = df_final.dropna(subset=['num_dept'])
```
